# Create output dataframe

```python
output = data.copy()
```

```python
## Remove data < 2020
output = output[output['YEAR'] >= 2020]
```

```python
## Create Date
output['DATE'] =output.apply(lambda x:str(int(x['DAY_OF_WEEK'])).zfill(2)+'/'+str(int(x['MONTH'])+1).zfill(2)+'/'+str(int(x['YEAR'])),axis=1)
output['DATE'].head(10)
```

```python
# Create low and High
from datetime import datetime
output['LOW'] =output['DEP_TIME_BLK'].apply(lambda x:x.split("-")[0])
output['HIGH'] =output['DEP_TIME_BLK'].apply(lambda x:x.split("-")[1])
output['TIMESTAMP'] = output.apply(lambda x: datetime.strptime(x['DATE']+' '+x['LOW'], '%d/%m/%Y %H%M').strftime("%Y/%m/%d %H:%M"),axis=1)
output.head(10)
```

```python
output['PRCP_SNOW_RATIO'] = output['PRCP'] / output['SNOW']
output['PLANE_AGE_AIRLINE_AIRPORT_FLIGHTS_MONTH_RATIO'] = output['PLANE_AGE'] / output['AIRLINE_AIRPORT_FLIGHTS_MONTH']
output['SEAT_DISTRIBUTION'] = output['NUMBER_OF_SEATS'] / output['CONCURRENT_FLIGHTS']


min_seat = output['SEAT_DISTRIBUTION'].min()
max_seat = output['SEAT_DISTRIBUTION'].max()
output['SEAT_DISTRIBUTION_NORMALISED'] = (output['SEAT_DISTRIBUTION'] - min_seat)/ (max_seat - min_seat)
```
