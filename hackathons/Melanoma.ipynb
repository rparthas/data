```python
import pandas as pd
import numpy as np
from pycaret.regression import *
```

```python
train = pd.read_csv('data/PM_Train.csv')
test = pd.read_csv('data/PM_Test.csv')
X_train = train.loc[:,train.columns != 'tumor_size']
y_train = train.loc[:,train.columns == 'tumor_size']
```

```python
reg1 = setup(data = train, target = 'tumor_size', normalize = True,
             normalize_method='zscore',feature_interaction = False, 
             feature_ratio=False,interaction_threshold=0.6,transformation=True,pca=True)
```

```python
model = compare_models()
```

```python
model = create_model('et')
```

```python
model_tuned = tune_model(model,optimize='RMSE') #3.9136 -> 3.8484 ->
```

```python
predict_model(model_tuned)
```

```python
final_model = finalize_model(model)
pred = predict_model(final_model, data=test)
pd.DataFrame(pred['Label']).rename(columns={'Label':'tumor_size'}).to_csv('data/submission.csv',index=False)
```
